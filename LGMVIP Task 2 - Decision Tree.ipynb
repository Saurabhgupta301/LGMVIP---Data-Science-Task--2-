# Performing Some Data Analysis

### Checking for null values...

```python
iris.isnull().sum()
```

### Displaying the info of iris dataset...

```python
iris.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()

iris.Species = lb.fit_transform(iris.Species)
iris
```

```python
iris.drop(columns = ['Id'],inplace = True)
```

```python
iris
```

## Seperating The Dataset into Features and Labels...

```python
x = iris.iloc[:,0:4]
y = iris.iloc[:,-1]
```

## Splitting the Dataset into train and test...

```python
# Splitting Data

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=0)
```

## Fitting the Model Using Decision Tree Algorithm...

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=5)

tree.fit(xtrain,ytrain)
```

## Performing Prediction...

```python
y_prediction = tree.predict(xtest)
```

## Checking the Accuracy of our Model...

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("The Accuracy of our Model is: {}".format(accuracy_score(ytest,y_prediction)))
```

#### *The Accuracy is Pretty Good :)*

## Let's Check where it got mistaken

```python
confusion_matrix(ytest,y_prediction)
```
